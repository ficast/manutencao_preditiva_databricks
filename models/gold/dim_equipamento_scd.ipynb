# Dimensão Equipamento SCD (Histórico)

Visão SCD Type 2 (Historical) baseada na Silver.

```python
spark.sql(f"""
CREATE OR REPLACE TABLE {fqtn(SCHEMA_GOLD, 'dim_equipamento_scd')} AS
SELECT 
    equipment_sk,
    equipment_id,
    equipment_name,
    status,
    location,
    effective_start,
    effective_end,
    is_current
FROM {fqtn(SCHEMA_SILVER, 'equipment_scd')}
""")
```
